```python
# Έστω ότι κάναμε "Retrieval" και βρήκαμε την εξής παράγραφο μέσα στο εταιρικό wiki:
retrieved_context = """
Πολιτική Αδειών 2026 (Ελληνικό Υποκατάστημα):
Όλοι οι εργαζόμενοι δικαιούνται 25 ημέρες κανονικής άδειας. 
Επιπλέον, προσφέρονται 5 ημέρες 'AI-Upskilling Leave' για παρακολούθηση σεμιναρίων.
Η άδεια πρέπει να εγκρίνεται από τον Manager 2 εβδομάδες νωρίτερα.
"""

user_question = "Πόσες μέρες άδειας συνολικά μπορώ να πάρω αν θέλω να πάω και σε ένα σεμινάριο AI;"

# Χτίζουμε το Agentic Prompt
messages = [
    SystemMessage(content=f"Είσαι ένας HR Assistant. Απάντα στην ερώτηση χρησιμοποιώντας ΑΥΣΤΗΡΑ το παρακάτω κείμενο:\n\n{retrieved_context}"),
    HumanMessage(content=user_question)
]

print("🤖 Σκέφτομαι...")
# response = chat(messages)
# print("\nΑπάντηση:", response.content)

# (Mock Output για το GitHub preview)
print("\nΑπάντηση: Μπορείτε να πάρετε συνολικά 30 ημέρες άδειας (25 ημέρες κανονικής άδειας + 5 ημέρες AI-Upskilling Leave). Θυμηθείτε να πάρετε έγκριση 2 εβδομάδες νωρίτερα!")
```
